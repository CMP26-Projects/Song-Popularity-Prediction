## Categorical EDA

```python
all_genres = dict()
for i in range(len(spotify_data['Genre'])):
    intermediate =spotify_data['Genre'][i].split("'") 
    if '[' in intermediate[:]:
        intermediate.remove('[')
    if ']' in intermediate[:]:
        intermediate.remove(']')
    if '[]' in intermediate[:]:
        intermediate.remove('[]')
    if ', ' in intermediate[:]:
        intermediate = list(filter(lambda a : a != ', ',intermediate))
    for word in intermediate:
        if word in all_genres:
            all_genres[word] = all_genres[word] + 1
        else:
            all_genres[word] = 1
```

```python
from itertools import islice

def take(n, my_dict):
    """Return the first n items of the iterable as a list."""
    return list(islice(my_dict, n))
```

```python
sorted_tuples =sorted(all_genres.items(), key=lambda x:x[1],reverse=True) 
sorted_dict = dict(sorted_tuples)
top20_genres = take(20, sorted_dict)
sorted_data = []
for i in top20_genres:
    sorted_data.append(all_genres[i])
plt.bar(top20_genres, sorted_data)
plt.xticks(rotation = 'vertical')
```

```python
genre_names = take(1545, all_genres)
streams_on_genre = dict()
for genre in genre_names:
    streams = 0
    for index, row in spotify_data.iterrows():
        if genre in row['Genre']:
            streams += row['Streams']
    streams_on_genre[genre] = streams
```

```python
sorted_tuples2 =sorted(streams_on_genre.items(), key=lambda x:x[1],reverse=True) 
sorted_dict2 = dict(sorted_tuples2)
all_genres = sorted_dict2.keys()
top20_genres2 = take(20, sorted_dict2)
sorted_data2 = []
for i in top20_genres2:
    sorted_data2.append(streams_on_genre[i])
plt.bar(top20_genres2, list(map(lambda x: x//(10**8), sorted_data2)))
plt.xticks(rotation = 'vertical')
```

```python
def map_chords(x):
    if x == "G#/Ab":
        return 1
    elif x == "A":
        return 2
    elif x == "D#/Eb":
        return 3
    elif x == "F#/Gb":
        return 4
    elif x == "E":
        return 5
    elif x == "F":
        return 6
    elif x == "B":
        return 7
    elif x == "C#/Db":
        return 8
    elif x == "A#/Bb":
        return 9
    elif x == "G":
        return 10
    elif x == "D":
        return 11
    else:
        return 12
```

```python
spotify_data['Chord'] = spotify_data['Chord'].map(lambda x : map_chords(x))
```

### Now Let's examine the above results graphically.

```python
def plot_line(x,y,axes,j,a,b):
    slope, intercept = np.polyfit(x, y, 1)

    # Generate y values for the regression line
    regression_line = slope * x + intercept

    # Plot the data points
    axes[j].scatter(x, y, label='Data Points')

    # Plot the regression line
    axes[j].plot(x, regression_line, color='red', label='Regression Line')

    # Add labels and title
    axes[j].set_xlabel(a)
    axes[j].set_ylabel(b[j])
    axes[j].set_title(a + " vs " + b[j])

    # Add a legend
    axes[j].legend()

    
```

```python
danceability_list = ['Energy','Acousticness','Valence','Loudness','Speechiness']
fig, axes = plt.subplots(len(danceability_list),figsize=(10,30))
j = 0
for i in danceability_list:
    plot_line(spotify_data['Danceability'],spotify_data[i],axes,j,'danceability',danceability_list)
    j += 1
    
```

```python
danceability_list = ['Acousticness','Valence','Loudness']
fig, axes = plt.subplots(len(danceability_list),figsize=(6,20))
j = 0
for i in danceability_list:
    plot_line(spotify_data['Energy'],spotify_data[i],axes,j,'Energy',danceability_list)
    j += 1
    
```

```python
valence_list = ['Artist Followers','Duration (ms)']
fig, axes = plt.subplots(len(valence_list),figsize=(6,12))
j = 0
for i in valence_list:
    plot_line(spotify_data['Valence'],spotify_data[i],axes,j,'Valence',valence_list)
    j += 1
    
```

### Question # : Is there a relation between the genre of a song and [Valence, Energy, Temp]?

```python
total_energy = []
total_temp = []
total_valence = []
for genre in sorted_dict2:
    energy_ = 0
    tempo_ = 0
    valence_ = 0
    for index, row in spotify_data.iterrows():
        if genre in row['Genre']:
            energy_ += row['Energy']
            tempo_ += row['Tempo']
            valence_ += row['Valence']
    total_energy.append(energy_)
    total_temp.append(tempo_)
    total_valence.append(valence_)
```

```python
total_energy = list(np.array(total_energy) / len(total_energy))
total_valence = list(np.array(total_valence) / len(total_valence))
total_temp = list(np.array(total_temp) / len(total_temp))
```

```python
fig, axes  = plt.subplots(3,figsize = (10,10))

axes[0].bar(top20_genres2,total_energy[0:20])
axes[0].set_xlabel("genres")
axes[0].set_ylabel("total_energy")

axes[1].bar(genre_names[0:20],total_valence[0:20])
axes[1].set_xlabel("genres")
axes[1].set_ylabel("total_valence")

axes[2].bar(genre_names[0:20],total_temp[0:20])
axes[2].set_xlabel("genres")
axes[2].set_ylabel("total_temp")
```
